```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
indep=dataset[['age','bp','al','su','bgr','bu','sc','sod','pot','hrmo','pcv','wc','rc','sg_b','sg_c','sg_d','sg_e','rbc_normal','pc_normal','pcc_present','ba_present','htn_yes','dm_yes','cad_yes','appet_yes','pe_yes','ane_yes',]]
dep=dataset[["classification_yes"]]
```

```python
indep
```

```python
dep
```

```python
dataset["classification_yes"].value_counts()
```

```python
indep.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion':['gini','entropy'],'max_features': ['auto','sqrt','log2'],'splitter':['best','random']} 
grid = GridSearchCV(DecisionTreeClassifier(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train, y_train)
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, grid_predictions)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, grid_predictions)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```
